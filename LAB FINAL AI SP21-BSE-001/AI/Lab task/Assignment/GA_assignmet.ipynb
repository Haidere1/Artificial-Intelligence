#### Introduction to GA
Genetic Algorithm (GA) is a search-based optimization technique based on the principles of Genetics and Natural Selection. It is frequently used to find optimal or near-optimal solutions to difficult problems which otherwise would take a lifetime to solve. It is frequently used to solve optimization problems, in research, and in machine learning.
 1. Genetic algorithms (GA) is a variant of stochastic beam search.
 2. Successor states are generated by combining two parents rather by modifying a single state. The process to generate new individauls form the parents is called crossover.
 3. The process is inspired by natural selection.
 4. Starts with $k$ randomly generated states, called population. Each state is an individual.
 5. An individual is usually represented by a string of 0’s and 1’s, or digits, or a finite set called chromosome.
 6. The objective function is called fitness function: better states have high values of fitness function.
 7. Pairs of individuals are selected at random for reproduction w.r.t. some probabilities.
 8. A crossover point is chosen randomly in the string.
 9. Offspring are created by crossing the parents at the crossover point.
 10. Each element in the string is also subject to some mutation with a small probability.

In our implementation a population $p$ will be a list of list. Let $p=[[c1],[c2]]$ be a population containing two chromosomes. Let $c1=[0,3,0,1]$ and $c2=[1,1,1,1]$ then the population can be given as $p=[[0,3,0,1],[1,1,1,1]]$. In the following function you have to generate an initital population. The  $chromosome_{length}$ means the number of genes in each chromosome. In the given example the chromosome $c1$ and $c2$ contain four genes, thus the $chromosome_{length}=4$. The $gene_{min}$ and $gene_{max}$ are the minimum and maximum values that a particular gene can have. For four queen the $gene_{min}$ and $gene_{max }$ may be 0 and 3 or 1 and 4 respectively. 

```python
def inital_population(p_size, chromosome_length, gene_min, gene_max):
    p=[] #initialize the population
    for i in range(p_size):
        c=[]
        for j in range(chromosome_length):
            c.append(rnd.randint(gene_min,gene_max))
        p.append(c)
    return p

#testing inital_population
p=inital_population(5,4,0,3)
print(p)
# For the given setup the output should be of the following form
#[[2, 3, 3, 0], [1, 2, 0, 2], [2, 1, 3, 2], [3, 1, 0, 3], [2, 1, 1, 0]]
        
```

### Selection
The basic part of the selection process is to stochastically select from one generation to create the basis of the next generation. The requirement is that the fittest individuals have a greater chance of survival than weaker ones. This replicates nature in that fitter individuals will tend to have a better probability of survival and will go forward to form the mating pool for the next generation. Weaker individuals are not without a chance. In nature such individuals may have genetic coding that may prove useful to future generations. The fitness values are computed using the fitness function (number of non-attacking pairs of queen) given above. We can see from the table (column Fitness) that individual No. 1 is the fittest and No. 4 is the weakest. Summing these fitness values we can apportion a percentage total of fitness. This gives the strongest individual a value of 31% and the weakest 14%.

These percentage fitness values can then be used to configure the roulette wheel. The number of times the roulette wheel is spun is equal to size of the population. As can be seen from the way the wheel is now divided, each time the wheel stops this gives the fitter individuals the greatest chance of being selected for the next generation and subsequent mating pool.

What is possibly more interesting from this example is that as the generations progress and the population gets fitter the gene pattern for individual No. 1 will become more prevalent in the general population because it is fitter, more apt to the environment we have put it in.

The selection can be performed in the following way:
- Compute the fitness of each chromosome of the population $p$
- Compute the \% fintess of each chromosome w.r.t the total fitness
- Find the commulative density of each chromosome (from first to the last)
- Generate a random number between 1 and 100. Select the chromosome in which range of accumulative \% fitness occurs the random number 

In the following function you have to implement the selection function.


```python
def selection(p):
    sum_fitness=0
    pf=[]# percent fitness
    for c in p:
        sum_fitness+=c[0]
    
    cpf=0 #cummolative fitness
    rn=rnd.randint(0,99) #Generate a random number between 0 and 100

    for c in p:
        cpf+=(c[0]*100)/sum_fitness
        if rn<=cpf:
            return c[1]
    return p[rnd.randint(0,len(p)-1)]
        
print(selection(p))

              
```

### Mutation:
Mutation is a genetic operator used to maintain genetic diversity from one generation of a population of genetic algorithm chromosomes to the next. It is analogous to biological mutation. Mutation alters one or more gene values in a chromosome from its initial state. In mutation, the solution may change entirely from the previous solution. Hence GA can come to a better solution by using mutation. Mutation occurs during evolution according to a user-definable mutation probability. This probability should be set low. If it is set too high, the search will turn into a primitive random search.

The classic example of a mutation operator involves a probability that an arbitrary bit in a genetic sequence will be changed from its original state. A common method of implementing the mutation operator involves generating a random variable for each bit in a sequence. This random variable tells whether or not a particular bit will be modified. This mutation procedure, based on the biological point mutation, is called single point mutation. Other types are inversion and floating point mutation. When the gene encoding is restrictive as in permutation problems, mutations are swaps, inversions, and scrambles.

Here, you have to do a single point mutation as follows:
- generate a random number $rn$
- if the $rn\le mp$ then change any one of the gene value. Here, $mp$, is the mutation probability. 
- Make sure that the new value of the gene lies in the range $[gene_{min}, gene_{max}]$

```python
def mutation(x,mp,gene_min,gene_max):
    rn=None # generate a random number [0,1]
    if rn<=mp:
        x[None]=None # Change the given gene value with a new gene value. Note that the new value should be in range [gene_min,gene_max]
    return x
        
```
